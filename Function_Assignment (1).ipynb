```python
#write a generator function in python that yields the powers of 2 up to a given exponent
def powers_of_two(max_exponent):
  """generator that yields powers of 2 up to 2^max_exponent."""
  for exponent in range(max_exponent + 1):
    yield 2 ** exponent
```

```python
#implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
  """generator that reads a file line by line and yields each line as a string."""
  with open(file_path,'r') as file:
    for line in file:
      yield line.strip()
```

```python
#use a lambda function in python to list a set of tuples based on the second element of each tuple.
tuples = [(1, 5), (3, 2), (2, 8), (4, 1)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])
print(sorted_tuples)
```
